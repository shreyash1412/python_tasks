Section B: Short Answer Questions
Answer in 1-2 sentences:

11. What is the difference between remove() and pop() methods in lists?
Answer: The remove() method removes the first occurrence of a specified value from a list, while the pop() method removes an element at a specific index and returns it.

12. What is the purpose of the pass statement in Python?
Answer: The pass statement is used as a placeholder for future code and is often used to create empty code blocks that do nothing.

13. Explain how the range() function works with an example.
Answer: The range() function generates a sequence of numbers within a specified range. 
For example, range(1, 5) will generate the numbers 1, 2, 3, and 4.

14. What is the difference between deep copy and shallow copy in Python?
Answer: A deep copy creates a new object and recursively copies all the objects it contains, while a shallow copy creates a new object and copies only the references to the original objects.

15. How do you check if a given key exists in a dictionary?
Answer: You can use the 'in' keyword to check if a given key exists in a dictionary. 
For example, 'key in dictionary' will return True if the key exists in the dictionary.

    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
